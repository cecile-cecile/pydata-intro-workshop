```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('data/bc-popular-girls-names.csv', index_col=0)
print(df.shape)
df.head()
```

```python
df.sort_values('Total', ascending=False).head(20)
```

```python
series = df.loc['JENNIFER'].drop('Total')
series.index = series.index.astype(int)
series.plot.line()
```

```python
yrly_totals = df.drop('Total', axis=1).sum(axis=0)
yrly_totals.index = yrly_totals.index.astype(int)
yrly_totals.head()
```

```python
def name_series_percent(df, yrly_totals, name):
    series = df.loc[name].drop('Total')
    series.index = series.index.astype(int)
    series =  100 * series / yrly_totals
    series.name = name
    return series
```

```python
jennifer = name_series_percent(df, yrly_totals, 'JENNIFER')
jennifer.describe()
```

```python
jennifer.name
```

```python
jennifer.plot.line()
plt.axvline(1981, color='k')
plt.title('JENNIFER')
```
